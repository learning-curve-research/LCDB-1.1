```python
import json
import glob
from collections import Counter

# Function to process each file
def process_files(file_pattern):
    error_messages = []
    
    # Get a list of all files matching the pattern
    files = glob.glob(file_pattern)
    
    # Iterate over each file
    for file_name in files:
        with open(file_name, 'r') as file:
            # Read file line by line
            for line in file:
                try:
                    # Deserialize the JSON object
                    json_obj = json.loads(line.strip())
                    # Check the value of the "status" key
                    if json_obj.get('status') != 'ok':
                        # Append the "error_message" to the list
                        error_message = json_obj.get('error_message')
                        if error_message:
                            error_messages.append(error_message)
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON in file {file_name}: {e}")
    return error_messages

# Pattern to match the files
file_pattern = 'results/result_*.json'

# Call the function to process the files and get error messages
error_messages = process_files(file_pattern)

# Count the unique error messages and their occurrences
error_message_counts = Counter(error_messages)

# Print the number of unique error messages and their occurrences
print(f"Number of unique error messages: {len(error_message_counts)}")
for error_message, count in error_message_counts.items():
    print(f"'{error_message}': {count} times")

```

```python
import json
import glob
from collections import defaultdict

# Function to process each file
def process_files(file_pattern):
    error_messages = defaultdict(lambda: {'count': 0, 'files': set()})
    
    # Get a list of all files matching the pattern
    files = glob.glob(file_pattern)
    
    # Iterate over each file
    for file_name in files:
        with open(file_name, 'r') as file:
            # Read file line by line
            for line in file:
                try:
                    # Deserialize the JSON object
                    json_obj = json.loads(line.strip())
                    # Check the value of the "status" key
                    if json_obj.get('status') != 'ok':
                        # Append the "error_message" to the dictionary
                        error_message = json_obj.get('error_message')
                        if error_message:
                            error_messages[error_message]['count'] += 1
                            error_messages[error_message]['files'].add(file_name.split('/')[-1])
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON in file {file_name}: {e}")
    
    # Convert sets to sorted lists for consistent output
    for error, data in error_messages.items():
        data['files'] = sorted(data['files'])
    
    return error_messages

# Pattern to match the files
file_pattern = 'results/result_*.json'

# Call the function to process the files and get error messages
error_messages = process_files(file_pattern)

# Print the number of unique error messages and their occurrences along with the files
print(f"Number of unique error messages: {len(error_messages)}")
for error_message, data in error_messages.items():
    files_str = ', '.join(data['files'])
    print(f"'{error_message}': {data['count']} times (files: {files_str})")
```
